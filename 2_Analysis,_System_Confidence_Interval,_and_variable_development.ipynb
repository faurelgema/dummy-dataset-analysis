**Analysis age of people using internet (get from birthdate)**

```python
X_train_data = df['age']
bins= [0,20,30,40,50,60]
labels = ['Late Ten','Twenty','Thirty','Fourty','Fifty']
df['age_group'] = pd.cut(X_train_data, bins=bins, labels=labels, right=False)
print(df)
```

let's add classification for NaN data

```python
df['age_group'] = df['age_group'].cat.add_categories('unknown').fillna('unknown')
```

```python
print(df)
```

```python
df.groupby(["age_group"])["id"].count().reset_index(name="total_people_by_age")
```

```python
df.age_group.value_counts().nlargest(5).plot(kind="bar", stacked=True, width=0.2)
```

there is unknown group of age that most using internet. and followed by late ten and twenty

# **System Confidence Interval**

```python
import numpy as np
import scipy.stats as st

data_test = df['age']
  
# create 90% confidence interval
st.t.interval(alpha=0.90, df=len(data_test)-1,
              loc=np.mean(data_test),
              scale=st.sem(data_test))
```

So with this system confidence result, I can take the conclusion, i really 90% confidence that mean from these ages from dataset have a range between 55 to 66.
