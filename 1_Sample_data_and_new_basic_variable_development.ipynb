```python
import pandas as pd
from pandas import to_datetime
from faker.providers import date_time
import random 
```

```python
from faker import Faker
```

```python
fake = Faker(['id_ID'])
type_of_device = ['ipad', 'iphone', 'android', 'mac', 'tablet']
digital_interest = ['ecommerce', 'travelling', 'music', 'science', 'social media']
location_type = ['ID']
data_to_merge = {
         'login_timestamp': [fake.date_time_this_year() for i in range(100)],
         'phone_number': [fake.phone_number() for i in range(100)]
         }

df_to_merge = pd.DataFrame(data_to_merge)
df_to_merge['type_of_device'] = [''.join(random.choice(type_of_device) for x in range(1)) for _ in range(len(df_to_merge))]
df_to_merge['digital_interest'] = [''.join(random.choice(digital_interest) for x in range(1)) for _ in range(len(df_to_merge))]
df_to_merge['location_type'] = [''.join(random.choice(location_type) for x in range(1)) for _ in range(len(df_to_merge))]
```

```python
df_to_merge
```

```python
fake = Faker(['id_ID'])
profiles = [fake.profile() for i in range(100)]
df = pd.DataFrame(profiles)
```

```python
df
```

```python
union_dfs = pd.concat([df, df_to_merge], axis=1)
print(union_dfs)
```

## **2. Checking for a new variabel**

a. create new variable **address**

```python
separator = ','
union_dfs[['address', 'city']] = union_dfs['residence'].str.split('\n', 1, expand=True)
union_dfs['city'] = [i.rsplit(separator, 1)[0] for i in union_dfs['city']]
```

b. create new variable **province**

```python
union_dfs['province'] = [i.rsplit(',', 1)[1] for i in union_dfs['residence']]
union_dfs['province'] = [i.rsplit(' ', 1)[0] for i in union_dfs['province']]
```

```python
union_dfs
```

c. Create new variable **age**

```python
from datetime import datetime
today = datetime.today()
union_dfs['age'] = union_dfs['birthdate'].apply(
               lambda x: today.year - x.year - 
               ((today.month, today.day) < (x.month, x.day)) 
               )
```
